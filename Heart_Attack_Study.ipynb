# Heart Attack Study

```python
# Dependences
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn.model_selection import train_test_split
```

```python
# Load the dataset
dataHeartAttacks = pd.read_csv('dataset-heartAttacks-ZheenHospital/Medicaldataset.csv', sep=",")

# The dataset contains 1319 rows
# Print the info on the dataset to check it was imported correctly
dataHeartAttacks.info()
```

```python
# Output only the header to have an overview of the data 
dataHeartAttacks.head()
```

```python
train_df, test_df = train_test_split(
    dataHeartAttacks,      # dataset
    test_size=0.20,        # 20 % testing â†’ 80 % training
    shuffle=True,          # randomize the data True 
    random_state=42,       # seed for repeatable results
    stratify= dataHeartAttacks["Result"]       # stratify the Result columns to balance classes
)
```

```python
train_df.head()
```

```python
test_df.head()
```

```python
# Convert the Result column in binary values 0 for negative and 1 for positive
train_df["Result"] = train_df["Result"].map({"negative": 0, "positive": 1})
train_df.head()
```

```python
# Do the same for the test dataframe
test_df["Result"] = test_df["Result"].map({"negative": 0, "positive": 1})
test_df.head()
```

```python
# Plot only the Age and Result column

test_df[0:10].plot(x='Age', y='Result', kind='scatter')
plt.title('Line Plot of X_values vs Y_values')
plt.xlabel('Age')
plt.ylabel('Heart Atack')
plt.grid(True)
plt.show()
```
